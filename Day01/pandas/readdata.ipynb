##  Reading data from Different Sources

```python
import pandas as pd
from io import StringIO

Data = '{"employee_name":"james","email":"james@gmail.com","job_profile":[{"developer":"team lead","seniority":"Sr. Developer"}]}'
df = pd.read_json(StringIO(Data))
```

```python
df
```

```python
## df to json
df.to_json()  ## orient = index by default
```

```python
df.to_json(orient='index')
```

```python
df.to_json(orient='records')  
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data?utm_source=chatgpt.com',header=None)
df.head()
```

```python
df.to_csv('iris.csv')
```

```python
url = 'https://www.fdic.gov/bank-failures/failed-bank-list' 
df = pd.read_html(url)

df = df[0]
df
```

```python
url = 'https://en.wikipedia.org/wiki/Mobile_country_code'

mobile_country_codes = pd.read_html(url,match='Country',header=0)
```

```python
import requests

url = 'https://en.wikipedia.org/wiki/Mobile_country_code'

# Send request with a browser-like header
html = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}).text

# Read tables from the HTML
df_country_codes= pd.read_html(html, match='Country', header=0)
```
